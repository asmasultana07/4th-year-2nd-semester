```python
st_length = int(input("Symbol Table Length: "))
```

```python
symbol_table={}
for i in range(st_length):
  symbol_table[i]={}
print(symbol_table)
```

```python
def getHashKey(name):
    ascii_value = 0

    for x in name:
        ascii_value += ord(x)  # ord(x) returns ascci value of a character

    hash_key = (ascii_value*len(name)) % st_length
    return hash_key
```

```python
def insert(name,tp):
  point=getHashKey(name)
  try:
    symbol_table[point][name]
    print("Data already exist.") 
  except:
    symbol_table[point][name]=tp
    print("Data upload successful")
```

```python
def search(name):
  try:
    point=getHashKey(name)
    symbol_table[point][name]
    print(f"Name: {name}")
    print("Type: ",symbol_table[point][name][0])
    print("Size: ", symbol_table[point][name][1])
    print("Dim: ",symbol_table[point][name][2])
    print("Line of Code: ",symbol_table[point][name][3])
    print("Line of Use: ",symbol_table[point][name][4])
    print("Address: ",symbol_table[point][name][5])
  except:
    print("Data not in Table")
```

```python
def update(name,new_name,tp):
  point= getHashKey(name)
  try:
    symbol_table[point][name]
    symbol_table[point].pop(name)
    new_point=getHashKey(new_name)
    symbol_table[new_point][new_name]=tp
  except:
    print("Data not found")
```

```python
def delete(name):
  point= getHashKey(name)
  try:
    symbol_table[point][name]
    symbol_table[point].pop(name)
    print("Delete Successful")
  except:
    print("Data not Found")
```

```python
def display():
  x= symbol_table.keys()
  

  for ky in x:
    y = symbol_table[ky].keys()
    for j in y:
      search(j)
      print("-----  -----")
```

```python
class SymbolTable:
    def __init__(self):
        self.table = []

    def insert(self, label, type_, size, dimension, line_of_code, address):
        self.table.append((label, type_, size, dimension, line_of_code, address))

    def search(self, label, type_):
        for entry in self.table:
            if entry[0] == label and entry[1] == type_:
                return entry
        return None

    def delete(self, label, type_):
        for entry in self.table:
            if entry[0] == label and entry[1] == type_:
                self.table.remove(entry)
                return True
        return False

    def update(self, label, type_, new_entry):
        for i, entry in enumerate(self.table):
            if entry[0] == label and entry[1] == type_:
                self.table[i] = new_entry
                return True
        return False

    def display(self):
        print("Label | Type | Size | Dimension | Line of Code | Address")
        print("-" * 55)
        for entry in self.table:
            print(" | ".join(entry))

def main():
    symbol_table = SymbolTable()

    while True:
        print("\nOptions:")
        print("1. Insert")
        print("2. Search")
        print("3. Delete")
        print("4. Update")
        print("5. Display")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            label = input("Enter Label: ")
            type_ = input("Enter Type: ")
            size = input("Enter Size: ")
            dimension = input("Enter Dimension: ")
            line_of_code = input("Enter Line of Code: ")
            address = input("Enter Address: ")
            symbol_table.insert(label, type_, size, dimension, line_of_code, address)

        elif choice == '2':
            label = input("Enter Label: ")
            type_ = input("Enter Type: ")
            result = symbol_table.search(label, type_)
            if result:
                print("Symbol found:", result)
            else:
                print("Symbol not found.")

        elif choice == '3':
            label = input("Enter Label: ")
            type_ = input("Enter Type: ")
            if symbol_table.delete(label, type_):
                print("Symbol deleted successfully.")
            else:
                print("Symbol not found.")

        elif choice == '4':
            label = input("Enter Label: ")
            type_ = input("Enter Type: ")
            new_entry = (label, type_, input("Enter Size: "), input("Enter Dimension: "), input("Enter Line of Code: "), input("Enter Address: "))
            if symbol_table.update(label, type_, new_entry):
                print("Symbol updated successfully.")
            else:
                print("Symbol not found.")

        elif choice == '5':
            symbol_table.display()

        elif choice == '6':
            print("Exiting...")
            break

        else:
            print("Invalid choice. Please choose again.")

if __name__ == "__main__":
    main()
```
